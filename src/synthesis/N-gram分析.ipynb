```python
import os
import glob
import collections
```

```python
def read_text_files(folder_path):
    """フォルダからテキストファイルを読み込み、テキストのリストを返す"""
    file_paths = glob.glob(os.path.join(folder_path, '*.txt'))
    texts = []
    for file_path in file_paths:
        with open(file_path, 'r', encoding='utf-8') as file:
            texts.append(file.read())
    return texts

def convert_katakana_to_hiragana(text):
    """テキスト中のカタカナをひらがなに変換する"""
    hira_text = text.translate(str.maketrans(
        "アイウエオカキクケコサシスセソタチツテトナニヌネノハヒフヘホマミムメモヤユヨラリルレロワヲンヴガギグゲゴザジズゼゾダヂヅデドバビブベボパピプペポァィゥェォャュョッ",
        "あいうえおかきくけこさしすせそたちつてとなにぬねのはひふへほまみむめもやゆよらりるれろわをんゔがぎぐげござじずぜぞだぢづでどばびぶべぼぱぴぷぺぽぁぃぅぇぉゃゅょっ"
    ))
    return hira_text

def tokenize_to_mora(text):
    """テキストをひらがなモーラに分割する"""
    yoon_list = ['ゃ', 'ゅ', 'ょ']
    hira_text = convert_katakana_to_hiragana(text)
    moras = []
    Fullmora = 0
    i = 0
    while i < len(hira_text):
        if i + 1 < len(hira_text) and hira_text[i + 1] in yoon_list:
            moras.append(hira_text[i] + hira_text[i + 1])
            i += 2
        else:
            moras.append(hira_text[i])
            i += 1

        Fullmora +=1
    return moras, Fullmora

def generate_ngrams(moras, n):
    """ひらがなモーラのリストからn-gramを生成する"""
    ngrams = [tuple(moras[i:i+n]) for i in range(len(moras)-n+1)]
    return ngrams

def count_ngrams(ngrams):
    """n-gramの出現数をカウントする"""
    return collections.Counter(ngrams)

def main(folder_path, n):
    countsAll = 0
    texts = read_text_files(folder_path)
    all_ngrams = []
    for text in texts:
        moras, counts = tokenize_to_mora(text)
        countsAll += counts
        
        ngrams = generate_ngrams(moras, n)
        all_ngrams.extend(ngrams)
    ngram_counts = count_ngrams(all_ngrams)
    return ngram_counts, countsAll

if __name__ == '__main__':
    folder_path = 'List_Lyrics_kana'  # 歌詞テキストファイルが保存されているフォルダーのパス
    n = 1 # バイグラムの場合は2、トライグラムの場合は3
    ngram_counts, CountsEnd = main(folder_path, n)
    
    print(CountsEnd)
#     テキスtファイルへの書き出し
    output_file = 'histgram/monogram_hist.txt'
    f = open(output_file, 'a')
    for ngram, count in ngram_counts.most_common():
        # count = 1
        f.write(f'{ngram}: {count}\n')

        
    print("----Text Output Finish----")
```
